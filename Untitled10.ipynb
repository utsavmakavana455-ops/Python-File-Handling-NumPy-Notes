```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Step 1: Create and Write File


feedback = [
    "Python is fun to learn.",
    "I love practical sessions.",
    "Too many assignments!",
    "The teacher explains clearly.",
    "I enjoy coding challenges.",
    "Sometimes the class is too fast.",
    "Python makes programming easier.",
    "The class activities are interesting.",
    "I need more Python practice.",
    "Overall, the course is excellent."
]

with open("feedback.txt", "w") as file:
    for line in feedback:
        file.write(line + "\n")

print("feedback.txt created successfully.\n")
```

```python

# Step 2: Read and Display


with open("feedback.txt", "r") as file:
    lines = file.readlines()

print("----- Student Feedback -----")

print(f"\nTotal number of lines: {len(lines)}\n")

for i, line in enumerate(lines, start=1):
    print(f"{i}: {line.strip()}")
```

```python
# Step 3: Text Cleaning


all_words = []

for line in lines:

    # Convert to lowercase
    line = line.lower()

    # Remove punctuation
    line = line.replace(".", "")
    line = line.replace(",", "")
    line = line.replace("!", "")
    line = line.replace("?", "")

    # Split into words
    words = line.split()

    # Store words
    all_words.extend(words)

print("\n---- Cleaned Words ----")
print(all_words)
```

```python

# Step 4: Analysis


total_words = len(all_words)

unique_words = len(set(all_words))

word_count = Counter(all_words)

most_frequent_word = word_count.most_common(1)[0]

python_count = word_count["python"]
class_count = word_count["class"]

print("\n------ Analysis -----")

print("Total words:", total_words)

print("Unique words:", unique_words)

print("Most frequent word:", most_frequent_word[0])

print("Frequency:", most_frequent_word[1])

print("Occurrences of 'python':", python_count)

print("Occurrences of 'class':", class_count)
```

```python


# Step 5: Save Results


with open("feedback_summary.txt", "w") as file:

    file.write(f"Total feedback lines: {len(lines)}\n")
    file.write(f"Total words: {total_words}\n")
    file.write(f"Unique words: {unique_words}\n")
    file.write(f"Most frequent word: {most_frequent_word[0]}\n")

print("\nfeedback_summary.txt created successfully.")
```
